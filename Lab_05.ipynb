```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# Загрузка данных
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
# 0: 'malignant' (злокачественная), 1: 'benign' (доброкачественная)
```

```python
X.head()
```

```python
print(X.isnull().sum())
```

```python
# Трейн тест сплит
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
# Масштабирование признаков
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Инициализация и обучение модели
# C — параметр регуляризации (чем меньше, тем сильнее штраф за большие веса)
model = LogisticRegression(C=1.0, solver='lbfgs')
model.fit(X_train_scaled, y_train)
```

```python
# Предсказание
y_pred = model.predict(X_test_scaled)
```

```python
# Оценка результатов
print(f"Общая точность (Accuracy): {accuracy_score(y_test, y_pred):.4f}")
print("\nДетальный отчет по метрикам:")
print(classification_report(y_test, y_pred, target_names=data.target_names))
```

```python
# Интерпретация: какие признаки важнее всего?
weights = pd.Series(model.coef_[0], index=data.feature_names)
top_weights = weights.sort_values(ascending=False).head(10)
print("\nТоп-10 признаков, влияющих на решение о доброкачественности:")
print(top_weights)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Загружаем датасет с рукописными цифрами
digits = load_digits()
X = digits.data  # Матрица признаков (64 пикселя на каждое изображение 8x8)
y = digits.target
```

```python
# Превращаем задачу в бинарную:
# Пусть 1 — это цифра '3', а 0 — любая другая цифра
y_binary = (y == 3).astype(int)

print(f"Всего объектов: {len(X)}")
print(f"Всего цифр '3': {sum(y_binary)}")
```

```python
# Трейн тест сплит
X_train, X_test, y_train, y_test = train_test_split(
    X, y_binary, test_size=0.3, random_state=42, stratify=y_binary
)
```

```python
# Линейным моделям необходимо масштабирование пикселей (от 0..16 к -1..1)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Обучаем логистическую регрессию
# class_weight='balanced' важен, так как "пятерок" в 9 раз меньше, чем остальных цифр
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
# Оценка
y_pred = model.predict(X_test_scaled)

print("Отчет о классификации:")
print(classification_report(y_test, y_pred, target_names=['Не 3', '3']))
```

```python
# 7. Посмотрим на веса (какие пиксели важнее всего для распознавания цифры 3)
# Мы можем восстановить форму 8x8 из весов модели
coef_image = model.coef_.reshape(8, 8)
print("\nМатрица весов (влияние пикселей):")
print(np.round(coef_image, 1))
```

# Задание
Поработать с датасетом breast cancer и используя лог. регрессию, l1, l2 регуляризации попытаться максимизировать recall для метки/класса malignant

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
"""
Постановка задачи - поиск лучшей модели для выявления рака груди.
1. Подготовка и масштабирование.
2. Перебор параметров с l1, l2. Алгоритмы liblinear, lbfgs.
"""


data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
best_recall = 0 # здесь храним лучший рез.
best_params = {} # лучшие параметры
best_model = None # лучшая модель

C_values = [0.001, 0.01, 0.1, 1, 10, 100]
penalties = ['l1', 'l2'] # типы регуляризации
solvers_l1 = ['liblinear'] 
solvers_l2 = ['lbfgs'] 

for penalty in penalties:
    if penalty == 'l1':
        solvers = solvers_l1
    else:
        solvers = solvers_l2
    
    for solver in solvers:
        for C in C_values:
            model = LogisticRegression(penalty=penalty, C=C, solver=solver, class_weight='balanced') # class_weight для корректировки веса классов
            model.fit(X_train_scaled, y_train)
            
            y_pred = model.predict(X_test_scaled)
            recall = recall_score(y_test, y_pred, pos_label=0)
            
            print(f"penalty={penalty} \t solver={solver} \t C={C} \t recall={recall:.6f}")
        
            # ищем лучшую модель
            if recall > best_recall:
                best_recall = recall
                best_params = {'penalty': penalty, 'solver': solver, 'C': C}
                best_model = model

print(f"\nЛучшие параметры:: {best_params}")
print(f"Лучший recall: {best_recall:.4f}")
```
